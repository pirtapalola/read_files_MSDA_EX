```python
#Combine the data from multiple DAT files to a single excel file.

os.chdir(r'C:/Users/username/data')

the_list = []
list1 = [i for i in range(1, 192)]
keys = np.array(list1)



for root, dirs, files in os.walk(r'C:/Users/username/data'):
    for file in files:
        if file.endswith('.dat'):
            the_list.append(file)
            
def readdataframe(the_list):
    df = pd.DataFrame() #define df as an empty pandas DataFrame
    for element in the_list:
        #print(element)
        df[element] = pd.read_csv(element, sep=' ', names=['TIA01_MGT_a'], header=0, usecols=[2], skiprows=[0])
    return df

trios_ramses = readdataframe(the_list)
#print(trios_ramses)
    
data1 = trios_ramses.iloc[43:234, file_list]
data2 = data1.astype('float64', copy=True, errors='raise')
spectrum = data2.divide(10)
spectral = spectrum.set_index(keys1, drop=True, append=False, inplace=False, verify_integrity=False)
print(spectral)

with pd.ExcelWriter('C:/Users/username/data/output/output.xlsx', mode='w+') as writer:  
    spectral.to_excel(writer)
```
